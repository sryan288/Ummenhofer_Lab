```python
import xarray as xr
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import cmocean
```

MHW detection has been done with the xmhw python package created by Paola Petrelli. The xmhw.detect function derives characteristics of detected events as well. The detection has been performed on daily values, spatially averaged over the shelf region based on the reference period of 1980 - 2019. Characteristics are based on the definition from Hobday et al. 2016.

```python
data = xr.open_dataset('https://data.geomar.de/thredds/dodsC/20.500.12085/b61b44fe-f4c8-4834-9c7f-91a5a9363b25/submitted/data/Fig_7/VIKING_MHW_events_shelf_1958_2019.nc').to_dataframe()

# setting array for intensity xticks
deg_ticks = np.arange(1,9,1)

# create seaborn joint grid plots for scatter plots of characteristic against depth and distribution of both; coloring of dots and bars according to the depth level
# duration
g = sns.JointGrid(data=data,x='duration',y='deptht',hue='depth',palette=cmocean.cm.dense,marginal_ticks=True,ratio=2)
g.plot_joint(sns.scatterplot,legend=False)
g.plot_marginals(sns.histplot,bins=30)

# set labels, ticks and panel numeration
g.fig.axes[0].invert_yaxis()
g.ax_joint.set_ylabel('depth [m]')
g.ax_joint.set_xlabel('duration [days]')
g.ax_marg_x.set_ylabel('count',visible=True)
g.ax_marg_y.set_xlabel('count',visible=True)
g.ax_marg_x.text(450,70,'A',fontsize=16,weight='bold')
g.ax_joint.text(450,15,'D',fontsize=16,weight='bold')
g.ax_marg_x.set_yticks([0,20,40,60,80])
g.ax_marg_x.set_yticklabels([0,20,40,60,80])
g.ax_marg_y.set_xticks([0,30,60,90])
g.ax_marg_y.set_xticklabels([0,30,60,90])
g.ax_joint.set_xticks([0,100,200,300,400])
g.ax_joint.set_xticklabels([0,100,200,300,400])

# MHW mean intensity
h = sns.JointGrid(data=data,x='intensity_mean',y='deptht',hue='depth',palette=cmocean.cm.dense,marginal_ticks=True,ratio=2)
h.plot_joint(sns.scatterplot,legend=False)
h.plot_marginals(sns.histplot,bins=30)

h.fig.axes[0].invert_yaxis()
h.ax_joint.set_xlabel(r'mean intensity [$^\circ$C]')
h.ax_joint.set_ylabel('')
h.ax_marg_x.set_ylabel('count',visible=False)
h.ax_marg_y.set_xlabel('count',visible=False)
h.ax_marg_x.text(7.5,25,'B',fontsize=16,weight='bold')
h.ax_joint.text(7.5,15,'E',fontsize=16,weight='bold')
h.ax_marg_x.set_yticks([0,10,20,30])
h.ax_marg_x.set_yticklabels([0,10,20,30])
h.ax_marg_y.set_xticks([0,30,60,90])
h.ax_marg_y.set_xticklabels([0,30,60,90])
h.ax_joint.set_xticks(deg_ticks)
h.ax_joint.set_xticklabels(deg_ticks)
h.ax_marg_x.set_xticks(deg_ticks)
h.ax_marg_x.set_xticklabels(deg_ticks)

# MHW maximum intensity
m = sns.JointGrid(data=data,x='intensity_max',y='deptht',hue='depth',palette=cmocean.cm.dense,marginal_ticks=True,ratio=2)
m.plot_joint(sns.scatterplot,legend=False)
m.plot_marginals(sns.histplot,bins=30)

m.fig.axes[0].invert_yaxis()
m.ax_joint.set_xlabel(r'max intensity [$^\circ$C]')
m.ax_joint.set_ylabel('')
m.ax_marg_x.set_ylabel('count',visible=False)
m.ax_marg_y.set_xlabel('count',visible=True)
m.ax_marg_x.text(7.5,25,'C',fontsize=16,weight='bold')
m.ax_marg_y.text(107.5,15,'G',fontsize=16,weight='bold')
m.ax_joint.text(7.5,15,'F',fontsize=16,weight='bold')
m.ax_marg_x.set_yticks([0,10,20,30])
m.ax_marg_x.set_yticklabels([0,10,20,30])
m.ax_marg_y.set_xticks([0,30,60,90])
m.ax_marg_y.set_xticklabels([0,30,60,90])
m.ax_joint.set_xticks(deg_ticks)
m.ax_joint.set_xticklabels(deg_ticks)
m.ax_marg_x.set_xticks(deg_ticks)
m.ax_marg_x.set_xticklabels(deg_ticks)
```

The three panels have been combined to one figure using inkscape
